# Read out data base
This notebook is an example for accessing the wave function database. 

```python
%load_ext autoreload
%autoreload 2
import sys
sys.path.append("../src") 
from nucwf import access
# silent excessive warnings  
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
# set pandas columns to no limit 
import pandas as pd 
pd.set_option('display.max_columns', None)
```

```python
# connect to standard data base for wave functions 
wfdb=access.database(workdir="/Users/andreasnogga/work/wflocal")
```

```python
# show current wave functions available 
wfdb.pddf
```

```python
# now download all files 
for index,row in wfdb.pddf.iterrows():
    # reformat row to dictionary
    rowdict=row.to_dict()    
    hashname,uniquefilename=wfdb.get_file(**rowdict)
    print('Obtained: ',hashname)
    
# now read some selected files 

# this is a A=3 file (see above)
hashname,uniquefilename=wfdb.get_file(**wfdb.pddf.loc[7].to_dict())
wavein=access.wavefa3(hashname,printlevel=1)

# this is a A=4 file (see above)
hashname,uniquefilename=wfdb.get_file(**wfdb.pddf.loc[9].to_dict())
wavein=access.wavefa4(hashname,printlevel=1)
```
